## `merge` (두 개 이상의 데이터프레임을 join, 연결)

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1 # 고객 목록
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2 # 구매 기록
```

```python
# merge : 결합하다 -> pandas.merge(df1, df2) # df1 (left), df2 (right)
# merge 와 concatnate 의 차이를 알아야한다.
pd.merge(df1,df2)

## df1, df2 합성 -> '고객번호'라는 공통의 열 항목으로 두 데이터프레임이 합쳐짐
```

```python
pd.merge(df2,df1) # df1, df2의 순서가 바뀐다면
                  # 열의 순서만 틀릴 뿐, 공통적으로 포함된 열은 자동으로 찾는다.
                  # 똑같이 공통된 열을 기준으로 합쳐진다.
```

```python
pd.merge(left=df1, right=df2, on='고객번호')  # -> 이렇게 명시적으로 표현해야했지만
                                              # 인수 위치에 맞춰서 입력만 하더라도 자동으로 공통열을 찾고
                                              # left right 구분해서 알아서 합쳐준다.
```

```python
pd.merge(left=df1, right=df2, on='고객번호', how = 'inner')
# inner join : (기준되는 열) 양쪽 데이터프레임에 모두 속하는 원소들만 남겨줌 (결측치 없는 nan이 안섞인...)
```

```python
pd.merge(left=df1, right=df2, on='고객번호', how = 'outer')
# 왼쪽, 오른쪽 안가리고 , 고객번호(에 대응되는)에 값이 있다면 다 가져온다.
# 아래 결과처럼 없는 데이터는 NaN 처리하여 다 가져와서 합침
```

[JOIN 관련 개념 이해 링크](https://hongong.hanbit.co.kr/sql-%EA%B8%B0%EB%B3%B8-%EB%AC%B8%EB%B2%95-joininner-outer-cross-self-join/)

```python
pd.merge(left=df1, right=df2, on='고객번호', how = 'left')
# how = left , 왼쪽에 있는 데이터들을 다 살려서 합친다.
# <-> how = right, 오른쪽에 있는 데이터들을 다 살려서 합친다.
```

```python
pd.merge(left=df1, right=df2, on='고객번호', how = 'right')
```

```python
df1 = pd.DataFrame({
    '품종': ['setosa', 'setosa', 'virginica', 'virginica'],
    '꽃잎길이': [1.4, 1.3, 1.5, 1.3]},
    columns=['품종', '꽃잎길이'])
df1
```

```python
df2.품종.value_counts()
```

```python
pd.merge(df1, df2)  # -> pd.merge(df1,df2, how='inner') 의 how ='inner'가 default.
# 1. 공통으로 존재하는 열을 찾음 (품종)
# 2. 품종들이 몇개씩 있는지 -> 같은 열에 동일한 원소가 여러 개
# -> 원소들끼리의 모든 경우의 수 (조합)을 만들어냄
# 3. inner (df1, df2가 모두 존재하는 걸로만 연결)
```

```python
pd.merge(df1, df2, how='outer') # 없는 데이터도 모두 합쳐서 보고 싶을 때, outer JOIN
```

### 연결되면 안되는 열의 이름이 같을 때

```python
# 두 df에서 이름이 같은 열은 모두 연결하는 기준 (키=on)
# 이름은 같아도 키가 되면 안되는 열 => `on` 무엇을 기준을 할지 명시
# (학생) 수업ID / 이름 (학생의 이름)
# (수업) 수업ID / 이름 (수업의 이름)
# -> on=[수업ID, 이름] ??? -> on='수업ID'

df1 = pd.DataFrame({
    '고객명': ['춘향', '춘향', '몽룡'],
    '날짜': ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터': ['20000', '30000', '100000']})
df1
```

```python
pd.merge(df1,df2) # -> pd.merge(df1,df2, on = ['고객명' , '데이터'])
# 두 데이터프레임 사이에 공통되는 열 이름이 '고객명' , '데이터' 두 개를 자동으로 기준 삼아
# 합친 결과 -> 하나도 겹치는게 없어 아무것도 출력 안한 결과.
```

## 그룹분석
* 조건에 맞는 데이터가 하나 이상일 경우에 -> 데이터 그룹을 이룸 (피봇 테이블 형태로 표현을 X)
* 그룹의 특성을 보여주는 (통계치, 계산...) => 그룹 분석 (group analysis)
* group -> `groupby`

1. 분석하고자 하는 Series, df에 `groupby` 메소드 호출 -> 그룹화
1. 그룹화 -> 그룹 객체 -> 그룹 연산

### groupby 메소드
* 데이터를 그룹별로 분류하는 역할.
    * 열 또는 열의 리스트 (어떤 열을 기준으로 그룹을 넣어줄 것이냐
    * 행 인덱스
=> GroupBy (그룹화) => 그룹 함수 or 그룹 연산 => 특정한 값. 배열...

```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
```

```python
df2.info()
```

```python
groups = df2.groupby(df2['key1']) # 특정한 열
# groups = df2.groupby(df2.key1)
groups
```

```python
# groupby.groups(속성) : 그룹별 구성 현황 (행 인덱스)
groups.groups # key1의 값 : 행 인덱스 리스트
```

```python
groups.count()  # 데이터 개수를 세는 것, NaN 제외하고
```

```python
df3= df2.copy()
df3.iloc[1,1] = np.nan
df3
```

```python
df3.groupby(df3.key1).count()
```

```python
groups.size() # 그룹 데이터의 행의 개수
```

```python
# 평균
groups.mean()
```

```python
# 중앙값 (median)
groups.median()
```

```python
# 최대값
groups.max()
```

```python
# 최소값
groups.min()
```

```python
# 합계
groups.sum()
```

```python
# 일괄적으로 곱함
groups.prod()   # production, product
```

```python
# 표준편차, 분산, 중간값(50%)
print(groups.std())
print(groups.var())
print(groups.quantile(0.5))
# print(groups.percentile(50))  # 여기서 percentile 안씀 -> quantile()을 사용!
```

```python
print(groups.quantile(0.25))
print(groups.quantile(0.75))
```

### agg (aggregate)

```python
import seaborn as sns # 예시 데이터들을 불러올 수 있음
iris = sns.load_dataset('iris') # 붓꽃
iris # 붓꽃의 데이터
```

```python
# 각 붓꽃의 종별로 (열별로) 가장 큰 값과 가장 작은 값의 비율을 구해보기

def peak_to_peak_ratio(x):   # 특정 열을 기준으로 묶인 그룹들의 묶음
    return x.max() / x.min()

# sepal (꽃받침) / petal (꽃잎)
# iris.groupby(iris.species).agg(lambda x: x.max() / x.min())

iris.groupby(iris.species).agg(peak_to_peak_ratio)
```

```python
iris.species.unique()
```

```python
iris.groupby(iris.species).agg(lambda x: x.max() / x.min())
```

* `agg` : 그룹 들의 <strong>대표값</strong>
* `apply` : 그룹들로 나뉘어진 행에 대해서 <strong>반복적 연산</strong>

```python
# 꽃잎의 길이가 가장 긴 상위 3개를 뽑는다
def top3_petal_length(df): # group
    # 내림차순 (큰게 위로...)
    return df.sort_values(by='petal_length', ascending=False)[:3] # 0, 1, 2
def top1_petal_length(df): # group
    # 내림차순 (큰게 위로...)
    return df.sort_values(by='petal_length', ascending=False)[:1] # 0, 1, 2

iris.groupby(iris.species).apply(top3_petal_length)
iris.groupby(iris.species).apply(top1_petal_length)
```
