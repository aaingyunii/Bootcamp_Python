### 그리드 설정 (grid)


```python
# 보조선 활성화/비활성화 : plt.grid(T/F)
def draw_line(grid=False, style='b-'):
    plt.title('나이와 재산의 상관관계')
    plt.plot(
        range(10, 50, 10), # x축 데이터
        [1, 4, 9, 16], # y축 데이터
        style # 차트의 스타일
    )
    # plt.xlabel('x축 라벨')
    plt.xlabel('나이')
    # plt.ylabel('y축 라벨')
    plt.ylabel('재산')

    # plt.grid(True)
    plt.grid(grid)

    plt.show()


draw_line()
draw_line(True)
draw_line(True, 'go:')
```

### 여러 개의 선 그어보기

```python
t= np.arange(0., 5., 0.2)
t
```

```python
# 1. 여러 개의 선 그리기 (하나의 plot 사용)
plt.title('라인 플롯에서 여러 개의 선#1')
plt.plot(
    t, t, 'r--',
    t, t ** 2, 'bs:',
    t, t ** 3, 'g^-',
)
plt.show()
```

```python
plt.title('라인 플롯에서 여러 개의 선#2')
charts = [t, t, 'r--',
          t, t ** 2, 'bs:',
          t, t ** 3, 'g^-',]
plt.plot(*charts)
plt.show()

plt.title('라인 플롯에서 여러 개의 선#3')
chart1 = [t, t, 'r--']
chart2 = [t, t ** 2, 'bs:']
chart3 = [t, t ** 3, 'g^-']
plt.plot(*chart1, *chart2, *chart3)
plt.show()
```

```python
# 2. 하나의 plot이 아니라 여러 개 -> 겹치는.
plt.title('여러 개의 plot 명령으로 한 그림에 겹쳐서 표현#1')

plt.plot((np.arange(4) + 1) ** 2)
plt.plot((np.arange(4) - 1) ** 3)

plt.show()
```

```python
plt.title('여러 개의 plot 명령으로 한 그림에 겹쳐서 표현#2')

for i in range(5):
    # plt.plot((np.arange(4) + 1) ** 2)
    # plt.plot((np.arange(4) - 1) ** 3)
    plt.plot((np.arange(4) + (1 if not i % 2 else -1)) ** (i+2))
plt.show()
```

```python
plt.title('여러 개의 plot 명령으로 한 그림에 겹쳐서 표현#3')

[plt.plot((np.arange(4) + (1 if not i % 2 else -1)) ** (i+2)) for i in range(3)]
plt.show()
```

```python
# 여러 개의 플롯이 겹쳤을 경우 라벨 표시
X = np.linspace(-np.pi, np.pi, 256)
X
```

```python
plt.title('삼각함수 plot')
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

#### `subplot`
* plot 명령을 여러 번 실행 => 같은 그래프 내부에 여러 직선 표시
* subplot : 행과 열 단위로 묶어서 여러 데이터를 병행해서 표시

```python
import numpy as np
```

```python
# 위아래 행 배치
# 행, 열, 서브플롯의 위치
plt.subplot(2, 1, 1)  # plt <- subplt(2,1,1) 지정해줄 수 있는...
plt.plot(np.arange(5) ** 2)
plt.title('차트1')

plt.subplot(2, 1, 2)
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title('차트2')
# 겹치는 문제
plt.tight_layout() # 안 겹치게 레이아웃 조정
plt.show()
```

```python
# 양옆 열 배치
# 행, 열, 서브플롯의 위치
plt.subplot(1, 2, 1)
plt.plot(np.arange(5) ** 2)
plt.title('차트1')

plt.subplot(1, 2, 2)
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title('차트2')

# 겹치는 문제
plt.tight_layout() # 안 겹치게 레이아웃 조정

plt.show()
```

```python
# 2x2 플롯 배치
plt.figure(figsize=(10,6))
plt.subplot(2, 2, 1)
plt.plot(np.arange(5) ** 2)
plt.title('차트1')
plt.xlabel('x축')
plt.ylabel('y축')

plt.subplot(2, 2, 2)
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title('차트2')

plt.subplot(2, 2, 3)
plt.plot(np.arange(5) ** 3, 'g-')
plt.title('차트3')

plt.subplot(2, 2, 4)
plt.plot(np.arange(10), np.arange(10) ** 2, 'b^-')
plt.title('차트4')

plt.tight_layout() # 안 겹치게 레이아웃 조정
plt.show()
```

### 다양한 차트 활용
* `lineplot` 선 차트
* 그외에 여러 가지 차트가 존재

```python
y = [2, 3, 1]
x = np.arange(3)

xlabel = ['가', '나', '다'] # len(x) 길이 같아야함
plt.title('바 차트')

plt.xticks(x, xlabel) # x축의 개별 값에다가 이름 붙여줌
plt.yticks(sorted(y)) # y축에다가 기준이 되는 이름을 붙여주기

plt.bar(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

y = [2, 3, 1]
x = np.arange(3)
xlabel = ['가', '나', '다']

colors = ['gold', 'silver', 'black']  # 각 값에 적용할 색상 지정

plt.title('바 차트')
plt.xticks(x, xlabel)

plt.bar(x, y, color=colors)  # 색상을 막대에 적용

plt.show()
```

```python
y = [2, 3, 1]
x = np.arange(3)
xlabel = ['가', '나', '다'] # len(x) 길이 같아야함
plt.title('가로 바 차트')
plt.yticks(x, xlabel) # x축의 개별 값에다가 이름 붙여줌
plt.xticks(sorted(y)) # y축에다가 기준이 되는 이름을 붙여주기
plt.barh(x, y) # -> barh (horizontal)
plt.show()
```

```python
people = ['james', 'john', 'kate', 'alice']
y_pos = np.arange(4)
performance = 3 + 10 * np.random.rand(4)
error = np.random.rand(4)

plt.title('Bar-h Chart')

plt.grid(True)
plt.barh(y_pos, performance, xerr=error, alpha=0.4) # alpha : 투명도 0.4 : 40%

plt.yticks(y_pos, people)
plt.xlabel('x축')
plt.show()
```

#### 파이 차트
* 원형의 파이 모양 그래프

```python
labels = ['호박파이', '사과파이', '정어리파이', '엄마손파이']

# sizes = [15, 30, 45, 10]
sizes = [15, 50, 72, 10]

# colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
colors = ['purple', 'gold', 'lightskyblue', 'lightcoral']

explode = (0, 0.1, 0.2, 0.3)  # 파이차트에서 튀어 나오게 하는 요소
plt.title("파이 차트")

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
# 각각의 파라미터 사용법을 아는 것이 중요 -> 다양한 모양으로 원하는 의미를 강조해서 표현이 가능.
# ...
# sizes -> 파이차트에서 항목이 차지하는 % 정도 / explode-> 부채꼴이 파이 차트의 중심에서 벗어나는 정도를 설정합니다.
# shadow-> True로 설정하면, 파이 차트에 그림자가 표시 / autopct -> 부채꼴 안에 표시될 숫자의 형식을 지정
# startangle는 부채꼴이 그려지는 시작 각도를 설정합니다.
#


plt.axis('equal')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# "FMRI" 데이터셋 로드
fmri = sns.load_dataset("fmri")

# 그래프를 그릴 subplot 설정
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# 첫 번째 subplot: region="parietal"인 경우의 시간에 따른 신호 평균 시각화
axes[0].plot(fmri.loc[fmri["region"] == "parietal", "timepoint"],
             fmri.loc[fmri["region"] == "parietal", "signal"])
axes[0].set_xlabel("Timepoint")
axes[0].set_ylabel("Signal")
axes[0].set_title("Signal in Parietal Region")

# 두 번째 subplot: region="frontal"인 경우의 시간에 따른 신호 평균 시각화
axes[1].plot(fmri.loc[fmri["region"] == "frontal", "timepoint"],
             fmri.loc[fmri["region"] == "frontal", "signal"])
axes[1].set_xlabel("Timepoint")
axes[1].set_ylabel("Signal")
axes[1].set_title("Signal in Frontal Region")

# subplot 간의 간격 조정
plt.tight_layout()

# 그래프를 표시
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# "FMRI" 데이터셋 로드
fmri = sns.load_dataset("fmri")

# 그래프를 그릴 subplot 설정
fig, axes = plt.subplots(figsize=(10, 8))

# region에 따른 시간에 따른 신호 평균 시각화
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="region", ax=axes)

# x축 레이블, y축 레이블, 그래프 제목 설정
axes.set_xlabel("Timepoint")
axes.set_ylabel("Signal")
axes.set_title("Average Signal by Timepoint")

# 범례 추가
axes.legend(title="Region")

# 그래프를 표시
plt.show()
```
