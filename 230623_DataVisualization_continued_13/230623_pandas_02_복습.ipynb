# Merge

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1 # 고객 목록
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2 # 구매 기록
```

```python
## df1, df2 -> 열의 항목에 차이가 있다. 고객번호의 구성 요소들이 다르다.
pd.merge(df1,df2) # -> pd.merge() : 두 데이터프레임 사이에서 '고객번호'라는 공통 열을 자동으로 찾아서
                  # 이를 기준으로 두 데이터프레임을 합쳤다.
                  # pandas.merge(left, right)

pd.merge(left=df1, right=df2, how = 'inner', on = '고객번호')
```

```python
pd.merge(left=df1, right=df2, how = 'outer', on = '고객번호') # 모든 데이터들을 가져와 합치겠다.
```

```python
pd.merge(left=df1, right=df2, how = 'left', on = '고객번호')  # left에 존재하는 모든행의 데이터 (기준 열),
                                                              # right에 NaN이 존재해도 그 상태로 가져오겠다.
```

```python
pd.merge(left=df1, right=df2, how = 'right', on = '고객번호') # right에 존재하는 모든행의 데이터 (기준 열),
                                                              # left에 NaN이 존재해도 그 상태로 가져오겠다.
```

```python
df1 = pd.DataFrame({
    '이름': ['영희', '철수', '철수'],
    '성적': [1, 2, 3]})
df1
```

```python
df2 = pd.DataFrame({
    '성명': ['영희', '영희', '철수'],
    '성적2': [4, 5, 6]})
df2
```

```python
# pd.merge(df1,df2) # MergeError: No common columns to perform merge on. Merge options: left_on=None, right_on=None, left_index=False, right_index=False
```

```python
pd.merge(df1,df2, left_on='이름', right_on='성명')
```

```python
pd.merge(df1, df2, left_on='이름', right_on='성명').drop('성명',axis=1)
# 원본에 영향을 안미치기 때문에, 사본을 만들기 때문
```

# JOIN

```python
df1 = pd.DataFrame({
    '품종': ['setosa', 'setosa', 'virginica', 'virginica'],
    '꽃잎길이': [1.4, 1.3, 1.5, 1.3]},
    columns=['품종', '꽃잎길이'])

df2 = pd.DataFrame({
    '품종': ['setosa', 'virginica', 'virginica', 'versicolor'],
    '꽃잎너비': [0.4, 0.3, 0.5, 0.3]},
    columns=['품종', '꽃잎너비'])
df1
```

```python
df2
```

```python
df1.join(df2, lsuffix = "l")  # 중복되는 열이 있을 경우 접미사를 통해서 변경
```

```python
df3 = df1.set_index('품종')
df3
```

```python
df4 = df2.set_index('품종')
df4
```

```python
df3.join(df4)
```

```python
pd.merge(df1,df2)
```

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]

df1 = pd.DataFrame(data, columns=columns)
df1
```
